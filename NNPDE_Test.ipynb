```python
PDE_name = 'AllenCahnEquation'

g_target_name = 'difficult'
```

```python
from PDESetting import PDESetting
from PDEsolvers import PDEsolver

PDESetting = PDESetting(PDE_name=PDE_name, g_target_name=g_target_name)
```

```python
import matplotlib.pyplot as plt

from IPython.display import clear_output, display

import numpy as np

epsilon = 0.32

g_target = PDESetting['g_target']
g_target_eval = g_target(PDESetting["tt"], PDESetting["xx"], PDESetting["yy"])

h = PDEsolver(PDESetting['u0'], g_target_eval, PDESetting["gamma"], epsilon, PDESetting["dt"], PDESetting["dx"], PDESetting["dy"], PDESetting["nt"], PDESetting["nx"], PDESetting["ny"])
nt = h.shape[0]

fig, ax = plt.subplots()
for n in np.arange(0, nt, min(nt, int(nt/200)+1)):
    #ax.imshow(h[i], origin='lower', aspect='auto')
    clear_output(wait=True)
    plt.imshow(h.detach().numpy()[n, :, :], cmap='viridis', origin='lower')
    plt.colorbar()
    plt.clim(vmin=h.min().item(), vmax=h.max().item())
    plt.pause(0.01)

```
